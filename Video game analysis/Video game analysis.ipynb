## Шаг 3. Проведите исследовательский анализ данных

### Сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

```python
#Сортировка таблицы по кол-ву игр в год
games_by_year = data.pivot_table(index='year_of_release', values='name', aggfunc='count')
games_by_year
```

```python
# Вывод гистограмы
plt.figure(figsize=(20, 7))
plt.bar(games_by_year.index, games_by_year['name'], color='b', alpha=0.7, width=0.8)
plt.xlabel('Год выпуска')
plt.ylabel('Кол-во выпущенных игр')
plt.grid()
plt.title(f'Кол-во выпущенных игр по годам')
plt.show()
```

На графике видно, что развитие игр с 1980 по 1995 года было слабым, так как индустрия только начала развиваться и широкого охвата игровыми приставками не было. С 1995 года был активный рост количества игр вместе с развитием новых технологий, веб сети, распространением персональных компьютеров с пиком в 2008 году. 

Также виден резкий спад с 2012, обусловленный экономическим кризисом и повышенными требованиям к процессу разработки игры и железу. Другой причиной спада является уход гейминга в онлайн, жизненый цикл игр значительно увеличился.

Считаю что данные за все периоды важны, благодаря этому можно проследить динамику разработки кол-ва игр.

```python
#выведем топ 10 платформ в сводную таблицу 
sales_by_platforms = data.pivot_table(index='platform', values='total_sales', aggfunc='sum')
sales_by_platforms = sales_by_platforms.sort_values(by='total_sales', ascending=False).head(10)
```

```python
sales_by_platforms
```

```python
# Зависимость кол-ва игр по годам для топ 10 платформ
(
    data
    .query('platform in @sales_by_platforms.index')
    .pivot_table(index='year_of_release', columns = 'platform', values='total_sales', aggfunc='sum')
    .plot(figsize = (20,7))
)
plt.show()
```

## Шаг 5. Проверьте гипотезы

### Одинаковые ли средние пользовательские рейтинги платформ Xbox One и PC

Сформулируем нулевую и анльтернативную гипотезы:

- Н0: Средний пользовательский рейтинг платформ Xbox One и PC равны
- H1: Средний пользовательский рейтинг платформ Xbox One и PC различается

```python
#введём 2 переменных с данными оценок по XOne и PC
XOne = data.query('platform == "XOne" and year_of_release >= 2013')['user_score'].dropna()
PC = data.query('platform == "PC" and year_of_release >= 2013')['user_score'].dropna()
```

```python
XOne, PC
```

```python
#зададим alpha
alpha = .05

#проведём тест
results = st.ttest_ind(XOne, PC, equal_var = False)

#выведем результат
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

На имеющихся данных, на уровне доверия 95% нет оснований отвергнуть нулевую гипотезу, средние пользовательские рейтинги платформ XOne и PC равны.

### Одинаковы ли средние пользовательские рейтинги жанров Action и Sports.

Сформулируем нулевую и анльтернативную гипотезы:

- Н0: Средние пользовательские рейтинги жанров Action и Sports равны.
- H1: Средние пользовательские рейтинги жанров Action и Sports разные.

```python
#введём 2 переменных с данными оценок по action и sports
action = data.query(' genre == "Action" ')['user_score'].dropna()
sports = data.query(' genre == "Sports" ')['user_score'].dropna()
```

```python
#зададим alpha
alpha = .05

#проведём тест
results = st.ttest_ind(action, sports, equal_var = False)

#выведем результат
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

На имеющихся данных, на уровне доверия 95% нет оснований отвергнуть нулевую гипотезу, средние пользовательские рейтинги жанров Action и Sports равны.
